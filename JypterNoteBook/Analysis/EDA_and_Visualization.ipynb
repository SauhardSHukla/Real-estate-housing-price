#### Import Lib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/5530/OneDrive/Desktop/Mall_Customer_Segmentation/Data_files/customer_arrivals_with_outliers.csv')

#-------------Basic Overview ------------------
eda_summary = df.describe(include='all')

#------Distribution of Numerical Feature--------
plt.Figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(df['Age'],kde=True,bins=20,color='skyblue')
plt.title("Distribution of Age")

plt.subplot(1, 2, 2)
sns.histplot(df['Total Spending'], kde=True, bins=20, color='salmon')
plt.title("Distribution of Total Spend")
plt.tight_layout()
plt.show()
```

```python
#--------Count plot for the gender----------
plt.figure(figsize=(8,4))

#This graph will show the all the values
sns.countplot(data=df,x='Gender',palette='Set2')
plt.title('Gender Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Count Plot for Days ---
plt.figure(figsize=(8, 4))

sns.countplot(data=df, x='Day', order=df['Day'].value_counts().index, palette='pastel')
plt.title("Customer Visits per Day")
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
